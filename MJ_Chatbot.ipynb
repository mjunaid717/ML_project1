```python
import gradio as gr
from langchain_openai import ChatOpenAI
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
import os

# Set OpenAI API Key (ensure this is securely managed in a real app)
os.environ["OPENAI_API_KEY"] = "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"

OPENROUTER_BASE_URL = "https://openrouter.ai/api/v1"
MODEL_NAME = "openai/gpt-oss-120b:free"

# Initialize LLM and Conversation Chain globally or with caching
# Using st.cache_resource equivalent for Gradio or just initializing once
llm_instance = ChatOpenAI(
    model=MODEL_NAME,
    temperature=0.7,
    base_url=OPENROUTER_BASE_URL,
)

# Initialize a new memory for each session or manage it within the chat function
def create_conversation_chain():
    memory = ConversationBufferMemory()
    conversation = ConversationChain(
        llm=llm_instance,
        verbose=False,
        memory=memory
    )
    return conversation

# Dictionary to hold conversation chains for different Gradio sessions
session_conversations = {}

def chat_interface(message, history):
    global session_conversations

    # Get a unique session ID for Gradio (e.g., from request, though not directly available here)
    # For simplicity, we'll use a fixed one for this example in Colab.
    # In a real deployed Gradio app, you might use gr.State or context info.
    session_id = 'default_gradio_session'

    if session_id not in session_conversations:
        session_conversations[session_id] = create_conversation_chain()

    conversation_chain = session_conversations[session_id]

    # Append current chat history to the conversation chain's memory
    # Note: Gradio's history format is List[List[str, str]]
    conversation_chain.memory.clear()
    for human_msg, ai_msg in history:
        conversation_chain.memory.save_context({"input": human_msg}, {"output": ai_msg})

    # Predict response using the conversation chain
    response = conversation_chain.predict(input=message)

    # Update history for the next turn
    conversation_chain.memory.save_context({"input": message}, {"output": response})

    return response


# Create the Gradio interface
gradio_interface = gr.ChatInterface(
    fn=chat_interface,
    title="Langchain Chatbot with Gradio",
    description="Ask me anything! I remember our conversation.",
    examples=[["Hi, how are you?"], ["What is the capital of France?"]],
    chatbot=gr.Chatbot(height=400)
)

# Launch the Gradio app
# To make it accessible externally in Colab, we use `share=True`
gradio_interface.launch(share=True)
```
