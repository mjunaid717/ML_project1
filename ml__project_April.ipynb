```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

ohe = OneHotEncoder()
ohe.fit(df[["Department"]])


ct = make_column_transformer((OneHotEncoder(categories=ohe.categories_),["Department"]), remainder = "passthrough")
x = ct.fit_transform(df.drop("Attrition", axis = 1))
y = df["Attrition"]

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
lr = LogisticRegression()
pipe = make_pipeline(lr)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import gradio as gr
import pandas as pd
import joblib
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

# Reload the dataset and apply preprocessing to set up ohe and ct
# NOTE: Original file '/content/Employee Attrition Data (1).csv' was not found.
# Please upload your file and update the path below. Using a placeholder for now.
# This will allow the Gradio app to launch, but predictions will not be meaningful
# until your actual data is loaded and the preprocessing steps below are uncommented and adjusted.
df_original = pd.read_csv("/content/sample_data/california_housing_train.csv")

# --- Temporarily added dummy columns to allow Gradio interface to launch ---
# These columns are expected by the Gradio interface inputs and the model preprocessing.
# They will not be meaningful for attrition prediction until actual data is loaded.
if 'Department' not in df_original.columns:
    df_original['Department'] = 'Dummy_Department'
if 'JobSatisfaction' not in df_original.columns:
    df_original['JobSatisfaction'] = 3
if 'OverTime' not in df_original.columns:
    df_original['OverTime'] = 0
if 'Age' not in df_original.columns:
    df_original['Age'] = 30
if 'Attrition' not in df_original.columns:
    df_original['Attrition'] = 0
# --- End of temporary dummy column additions ---

# NOTE: The following original data cleaning and preprocessing steps are commented out
# because they are specific to your original dataset and might cause errors or be irrelevant
# with the placeholder data.
# Uncomment and adjust these lines once you have uploaded your actual dataset.
# df_original.dropna(inplace=True)
# df_original = df_original[df_original['Age'].astype(str).str.isnumeric()]
# df_original = df_original[df_original['JobSatisfaction'].astype(str).str.isnumeric()]
# df_original["OverTime"].replace({"Y":1,"N":0, "Yes":1,"No":0, "yes":1,"no":0}, inplace = True)
# df_original["Attrition"].replace({"Y":1,"N":0, "Yes":1,"No":0, "yes":1,"no":0}, inplace = True)
# df_original['JobSatisfaction'] = df_original['JobSatisfaction'].astype(str).str.replace(",","").astype(int)
# df_original['Age'] = df_original['Age'].astype(str).str.replace(",","").astype(int)

# Define feature_cols, ensuring it includes the Gradio input features and any others
# that the original model was trained on (best guess for now).
# This list should ideally be the exact list of columns from df.drop("Attrition", axis=1) from training.
# For now, we'll try to infer it from the columns available after placeholder/dummy additions.
feature_cols = [col for col in df_original.columns if col not in ['Attrition']]

# Recreate OneHotEncoder and ColumnTransformer
ohe_gradio = OneHotEncoder(handle_unknown='ignore')
ohe_gradio.fit(df_original[["Department"]])

ct_gradio = make_column_transformer(
    (OneHotEncoder(categories=ohe_gradio.categories_, handle_unknown='ignore'), ["Department"]),
    remainder="passthrough"
)

# Fit the column transformer on the available feature columns
ct_gradio.fit(df_original[feature_cols])

# Load the trained logistic regression model
# Note: The model might expect a specific number of features after transformation.
# This might still cause an error during prediction if the placeholder data's transformed features
# don't match the model's expected input shape.
lr_model = joblib.load('logistic_regression_model.pkl')

# Calculate default values for features. These will be used in the predict_attrition function.
default_values = {}
for col in feature_cols:
    if df_original[col].dtype == 'object':
        default_values[col] = df_original[col].mode()[0]
    else:
        default_values[col] = df_original[col].mean()

# Ensure default values for the explicit Gradio inputs are set (can override the above if needed)
default_values['Age'] = 30
default_values['Department'] = 'Dummy_Department'
default_values['JobSatisfaction'] = 3
default_values['OverTime'] = 0
```

```python
def predict_attrition(age, department, job_satisfaction, over_time):
    # Create a DataFrame with all features, using defaults for most
    input_data = pd.DataFrame([default_values]) # Start with default values

    # Update with user provided values
    input_data['Age'] = age
    input_data['Department'] = department
    input_data['JobSatisfaction'] = job_satisfaction
    input_data['OverTime'] = 1 if over_time else 0 # Convert boolean to 0/1

    # Ensure column order matches the training data
    input_data = input_data[feature_cols] # Ensure the order of columns

    # Transform the input data using the fitted ColumnTransformer
    transformed_input = ct_gradio.transform(input_data)

    # Make prediction using the loaded model
    prediction = lr_model.predict(transformed_input)
    prediction_proba = lr_model.predict_proba(transformed_input)[:, 1] # Probability of attrition

    if prediction[0] == 1:
        return f"Prediction: Employee is likely to attrite (Probability: {prediction_proba[0]:.2f})"
    else:
        return f"Prediction: Employee is unlikely to attrite (Probability: {prediction_proba[0]:.2f})"

# Get unique department values for dropdown
department_choices = df_original['Department'].unique().tolist()

# Create Gradio Interface
interface = gr.Interface(
    fn=predict_attrition,
    inputs=[
        gr.Slider(minimum=18, maximum=65, step=1, value=30, label="Age"),
        gr.Dropdown(department_choices, label="Department", value=department_choices[0]),
        gr.Slider(minimum=1, maximum=4, step=1, value=3, label="Job Satisfaction"),
        gr.Checkbox(label="Over Time (worked over time)", value=False)
    ],
    outputs="text",
    title="Employee Attrition Prediction",
    description="Predict if an employee is likely to attrite based on key factors."
)

# Launch the interface
interface.launch(debug=True)
```
